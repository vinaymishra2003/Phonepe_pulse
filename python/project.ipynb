```python
#This is to direct the path to get the data as states

path="D:/project folder/first_project/data/aggregated/insurance/country/india/state/"
Agg_state_list=os.listdir(path)
Agg_state_list
```

```python
#This is to extract the data's to create a dataframe

clm={'State':[], 'Year':[],'Quater':[],'Transacion_type':[], 'Transacion_count':[], 'Transacion_amount':[]}

for i in Agg_state_list:
    p_i=path+i+"/"
    Agg_yr=os.listdir(p_i)
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']['transactionData']:
              Name=z['name']
              count=z['paymentInstruments'][0]['count']
              amount=z['paymentInstruments'][0]['amount']
              clm['Transacion_type'].append(Name)
              clm['Transacion_count'].append(count)
              clm['Transacion_amount'].append(amount)
              clm['State'].append(i)
              clm['Year'].append(j)
              clm['Quater'].append(int(k.strip('.json')))
#Succesfully created a dataframe
Agg_Insur=pd.DataFrame(clm)
```

```python
path="D:/project folder/first_project/data/aggregated/user/country/india/state/"
Agg_state_list=os.listdir(path)

clm = {
    'State': [],
    'Year': [],
    'Quarter': [],
    'Brand': [],
    'User_count': [],
    'Percentage': []
}

for i in Agg_state_list:
    p_i = path + i + "/"
    Agg_yr = os.listdir(p_i)

    for j in Agg_yr:
        p_j = p_i + j + "/"
        Agg_qtr_list = os.listdir(p_j)

        for k in Agg_qtr_list:
            p_k = p_j + k
            with open(p_k, 'r') as Data:
                D = json.load(Data)

            users_device = D.get('data', {}).get('usersByDevice')
            if users_device is None:
                continue

            quarter = int(k.replace('.json', ''))  # ✅ FIX

            for z in users_device:
                clm['Brand'].append(z['brand'])
                clm['User_count'].append(z['count'])
                clm['Percentage'].append(z['percentage'])
                clm['State'].append(i)
                clm['Year'].append(j)
                clm['Quarter'].append(quarter)

Agg_User = pd.DataFrame(clm)

```

```python
path="D:/project folder/first_project/data/map/insurance/hover/country/india/state/"
map_state_list=os.listdir(path)

clm = {
    'State': [],
    'Year': [],
    'Quarter': [],
    'District': [],
    'Transaction_type': [],
    'Transaction_count': [],
    'Transaction_amount': []
}

for i in map_state_list:
    p_i = path + i + "/"
    map_yr = os.listdir(p_i)

    for j in map_yr:
        p_j = p_i + j + "/"
        map_qtr_list = os.listdir(p_j)

        for k in map_qtr_list:
            p_k = p_j + k
            with open(p_k, 'r') as Data:
                D = json.load(Data)

            hover_list = D.get('data', {}).get('hoverDataList')
            if hover_list is None:
                continue

            quarter = int(k.replace('.json', ''))  # ✅ FIX

            for z in hover_list:
                district = z['name']

                for m in z['metric']:
                    clm['State'].append(i)
                    clm['Year'].append(j)
                    clm['Quarter'].append(quarter)
                    clm['District'].append(district)
                    clm['Transaction_type'].append(m['type'])
                    clm['Transaction_count'].append(m['count'])
                    clm['Transaction_amount'].append(m['amount'])

# Create DataFrame
map_Insur = pd.DataFrame(clm)

```

```python
path="D:/project folder/first_project/data/top/insurance/country/india/state/"
top_state_list=os.listdir(path)

clm = {
    'State': [],
    'Year': [],
    'Quarter': [],
    'Transaction_type': [],
    'Transaction_count': [],
    'Transaction_amount': []
}

for i in top_state_list:
    p_i = path + i + "/"
    top_yr = os.listdir(p_i)

    for j in top_yr:
        p_j = p_i + j + "/"
        top_qtr_list = os.listdir(p_j)

        for k in top_qtr_list:
            p_k = p_j + k
            with open(p_k, 'r') as Data:
                D = json.load(Data)

            data = D.get('data', {})
            quarter = int(k.replace('.json', ''))

            # Loop through states, districts, pincodes
            for level in ['states', 'districts', 'pincodes']:
                entity_list = data.get(level)
                if entity_list is None:
                    continue

                for item in entity_list:
                    metric = item['metric']

                    clm['State'].append(i)
                    clm['Year'].append(j)
                    clm['Quarter'].append(quarter)
                    clm['Transaction_type'].append(metric['type'])
                    clm['Transaction_count'].append(metric['count'])
                    clm['Transaction_amount'].append(metric['amount'])

# Create DataFrame
top_insurance = pd.DataFrame(clm)

```
